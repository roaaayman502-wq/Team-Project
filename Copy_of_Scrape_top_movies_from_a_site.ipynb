## 300 Best Movies of All Time

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://editorial.rottentomatoes.com/guide/best-movies-of-all-time/'
response = requests.get(url)
response.status_code
```

```python
soup = BeautifulSoup(response.text, 'html')
print(soup.prettify())
```

```python
scraped_movies = soup.find_all('a' , class_ = 'title')
movies = []
for x in scraped_movies:
  movies.append(x.text)
print(movies)
```

```python
scraped_years = soup.find_all('span',class_ ="year")
years = []
for x in scraped_years:
  years.append(x.text
              .replace('(' , '')
              .replace(')', '')
  )
print(years)
```

```python
scraped_scores = soup.find_all('span',class_="score")
scores = []
for x in scraped_scores:
  scores.append(x.text)
print(scores)
```

```python
min_len = min(len(movies), len(years), len(scores))

data = pd.DataFrame({
    "Movie": movies[:min_len],
    "Year": years[:min_len],
    "Score": scores[:min_len],
})

data.head(10)
```
