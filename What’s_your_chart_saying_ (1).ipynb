```python
import matplotlib.pyplot as plt

scores = [68, 74, 81, 77, 92, 85, 79]   # test scores
hours = [3, 4, 6, 5, 8, 7, 6]      # hours studied

subjects = ["Math", "Science", "English"]
averages = [88, 72, 80]    # subject averages
```

1. Bar Chart - Subject Averages

```python
plt.bar(subjects, averages, color="skyblue", edgecolor="black")
plt.title("Average Scores by Subject")
plt.xlabel("Subjects")
plt.ylabel("Average Score")
plt.show()
```

Story: “Math has the highest average score (88), which means students are doing well in this subject. Science, however, is clearly the weakest (72). English sits in the middle (80). This tells a story: if teachers want to improve class performance, Science may need extra focus or support.”

2. Histogram – Score Distribution

```python
plt.hist(scores, bins=5, color="orange", edgecolor="black")
plt.title("Score Distribution")
plt.xlabel("Score Ranges")
plt.ylabel("Number of Students")
plt.show()
```

Story: “Most students scored between 70 and 90. Only a few are below 70, and nobody scored extremely low. The histogram shows a cluster around the 70–85 range, meaning the class is performing steadily, with just a few standout high achievers.”

3. Scatter Plot –Study Hours vs Scores

```python
plt.scatter(hours, scores, color="purple")
plt.title("Study Hours vs Test Scores")
plt.xlabel("Hours Studied")
plt.ylabel("Scores")
plt.grid(True)
plt.show()
```

Story: “There’s a clear positive relationship: students who studied more hours generally scored higher. For example, 2 hours of study gave a score of 68, while 7–8 hours gave scores above 85. This supports the idea that effort (study time) pays off in higher performance.”
